# Data Exploration - Trade AI

Exploration de données Forex et Or

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from src.data.collector import DataCollectorFactory
from src.data.preprocessor import DataPreprocessor

# Configuration
sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = (15, 6)
```

```python
# Collecter les données
collector = DataCollectorFactory.create_collector('EURUSD')
df = await collector.fetch_historical_data('EURUSD', 'daily', 365)
print(f'Shape: {df.shape}')
print(df.head())
```

```python
# Prétraitement
preprocessor = DataPreprocessor()
df_clean = preprocessor.clean_data(df)
df_featured = preprocessor.add_technical_features(df_clean)
print(df_featured.columns.tolist())
```

```python
# Visualisation des prix
plt.figure()
plt.plot(df_featured.index, df_featured['close'], label='Close Price')
plt.plot(df_featured.index, df_featured['sma_20'], label='SMA 20')
plt.plot(df_featured.index, df_featured['sma_50'], label='SMA 50')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('EURUSD - Prix et Moyennes Mobiles')
plt.legend()
plt.show()
```

```python
# Indicateurs techniques
fig, axes = plt.subplots(3, 1, figsize=(15, 10))

# RSI
axes[0].plot(df_featured.index, df_featured['rsi'])
axes[0].axhline(y=30, color='r', linestyle='--', alpha=0.5)
axes[0].axhline(y=70, color='r', linestyle='--', alpha=0.5)
axes[0].set_ylabel('RSI')
axes[0].set_title('RSI (14)')

# MACD
axes[1].plot(df_featured.index, df_featured['macd'], label='MACD')
axes[1].plot(df_featured.index, df_featured['macd_signal'], label='Signal')
axes[1].set_ylabel('MACD')
axes[1].legend()

# Volume
axes[2].bar(df_featured.index, df_featured['volume'], alpha=0.5)
axes[2].set_ylabel('Volume')
plt.tight_layout()
plt.show()
```
